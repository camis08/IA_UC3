## 7. Tratamento de texto e padronização

### Exercício 13
Padronize a coluna `bolsista` para ficar apenas com:
- `Sim`
- `Não`

Dica: existem valores como `SIM`, `sim`, `Nao`.

### Exercício 14
Padronize a coluna `uf` para letras maiúsculas.

### Exercício 15
Padronize a coluna `cidade` no formato título.
Exemplo:
- `brasília` → `Brasília`


```python
df.info()
```

```python
# padronize a coluna bolsista para ficar apenas com:
"""
Sim
Não
"""
#Dica: existem valores como SIM, sim, Nao

"""
df['bolsista'] = df['bolsista'].str.capitalize()
df['bolsista']
"""

df["bolsista"] = df["bolsista"].str.strip().str.lower()
df["bolsista"] = df["bolsista"].replace({
    "sim": "Sim",
    "nao": "Não",
    "não": "Não"
})

df["bolsista"].unique()
```

```python
# padronize a coluna uf para letras maiúsculas.
df["uf"] = df["uf"].str.upper()
df["uf"]
```

```python
# padronize a coluna cidade no formato título. Exemplo:
"""
brasília → Brasília
"""
df["cidade"] = df["cidade"].str.title()
df["cidade"]
```

## 8. Criando novas colunas

### Exercício 16
Crie uma coluna chamada `media` com a média entre `nota1` e `nota2`.

### Exercício 17
Crie uma coluna chamada `aprovado`:
- `True` para média maior ou igual a **7**
- `False` caso contrário

### Exercício 18
Crie uma coluna chamada `situacao_faltas`:
- `Alto risco` para mais de 10 faltas
- `Atenção` para 6 a 10 faltas
- `Ok` para até 5 faltas


```python
# Crie uma coluna chamada media com a média entre nota1 e nota2
df["media"] = (df["nota1"] + df["nota2"]) / 2
df["media"].round(2)
```

```python
df
```

```python
# Crie uma coluna chamada aprovado:

"""
True para média maior ou igual a 7
False caso contrário
"""
df["aprovados"] = df["media"] >= 7
df["aprovados"]
```

```python
# Crie uma coluna chamada situacao_faltas:

"""
Alto risco para mais de 10 faltas
Atenção para 6 a 10 faltas
Ok para até 5 faltas
"""

"""
df["situacao_faltas"] = if df["faltas"] > 10: "Alto risco" elif df["faltas"] >= 6 and df["faltas"] <= 10: "Atenção" else: "Ok"
df["situacao_faltas"]
"""

def classificar_faltas(faltas):
    if faltas > 10:
        return "Alto risco"
    elif faltas >= 6:
        return "Atenção"
    else:
        return "Ok"

df["situacao_faltas"] = df["faltas"].apply(classificar_faltas)
df["situacao_faltas"]
```
