# Revisão: Blocos, Condições e laços em python

## 1. Blocos de código em Python

Em Python, blocos são definidos pela indentação.

```python
if True:
    print("Este comando está dentro do bloco")
    print("Este também está dentro")

print("Este está fora do bloco")
```

Observe que os comandos dentro do if estão indentados.

Normalmente usamos 4 espaços.

**Exercício**

Execute o código abaixo e observe o resultado.

```python
x = 10

if x > 5:
    print("x é maior que 5")

print("Fim do programa")
```

## 2. Estruturas Condicionais

Permitem que o programa tome decisões.

Estrutura básica:

```python
"""
if condição:
    código
"""
```

Exemplo

```python
idade = 20

if idade >= 18:
    print("Maior de idade")
```

Usando else

```python
idade = 16

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

Usando elif

```python
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

## 3. Laço for

```python
for i in range(5):
    print(i)
```

```python
for i in range(1, 11):
    print(i)
```

**Exercício**

Mostre a tabuada do 5.

```python
Exemplo de saída:
5 x 1 = 5
5 x 2 = 10
...
```

```python
# escreva aqui
for i in range(1,11):
  print(f"5 * {i} = {5*i}")
```

## 4. Laço while

O while executa enquanto uma condição for verdadeira.

```python
contador = 0

while contador < 5:
    print(contador)
    contador += 1
```

Exemplo

```python
numero = 1

while numero <= 10:
    print(numero)
    numero += 1
```

**Exercício**

Crie um while que mostre números de 10 até 1.

```python
# escreva aqui
numero = 10

while numero >= 1:
    print(numero)
    numero -= 1
```

## 5. Controle de fluxo

**Break**

Interrompe o loop.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

**Continue**

Pula uma iteração.

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

## 6. Laços com listas

```python
frutas = ["maçã", "banana", "laranja"]

for fruta in frutas:
    print(fruta)
```

**Exercício**

Crie uma lista com 5 números.

Use um for para mostrar o quadrado de cada número.

```python
"""
Exemplo:
2 → 4
3 → 9
"""
```

```python
numeros = [8,18,26,22,32]

for num in numeros:
  print(f"Numeros : {num} → {num**2}")
```

## 7. Exemplo prático

Somar números de uma lista.

```python
numeros = [10, 20, 30, 40]

soma = 0

for n in numeros:
    soma += n

print("Soma:", soma)
```

**Exercício**

Calcule a média de uma lista de números.
Essa lista pode ter qualquer tamanho e qualquer número. Seu código tem que identificar o tamanho da lista automaticamente e calcular a média dos valores presentes.

```python
numeros = [3,10,21,22,50,58]
media = sum(numeros) /len(numeros)
print(f"{media:.2f}")
```
