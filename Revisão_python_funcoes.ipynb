**O que são funções?**

Funções permitem reutilizar código.

```python
def saudacao():
    print("Olá aluno!")
```

```python
saudacao()
```

### Funções com parâmetros

```python
def saudacao(nome):
    print("Olá", nome)
```

```python
saudacao("Maria")
```

### Funções com retorno

```python
def soma(a, b):
    return a + b
```

```python
resultado = soma(10, 5)
print(resultado)
```

```python
soma(10, 5)
```

**EXERCÍCIO 05:**

 Crie uma função soma_imposto(taxa_imposto, custo) que altere o valor do custo incluindo o imposto (ex: taxa de 10% sobre 100 retorna 110)

```python
def soma_imposto(taxa_imposto,custo):
  porcentagem = taxa_imposto / 100
  imposto = custo * porcentagem
  return custo + imposto
  #return(custo + custo*(taxa_imposto /100)

print(soma_imposto(10,100))
#soma_imposto(10,100)
```
