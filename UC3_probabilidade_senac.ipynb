# Princípio Fundamental da Contagem (PFC)

##  O que é?

O Princípio Fundamental da Contagem diz:

> Se um processo ocorre em etapas independentes, multiplicamos as possibilidades de cada etapa.

---

## Intuição

Se você tem escolhas em sequência, você **multiplica**.

---



##  Exemplo 1 — Roupas

- 3 camisetas  
- 2 calças  

Quantos looks diferentes?

$$
3 \times 2 = 6
$$

---



##  Exemplo 2 — Moeda + Dado

- Moeda → 2 opções (cara, coroa)  
- Dado → 6 opções  

Total de possibilidades:

$$
2 \times 6 = 12
$$

---



##  Conexão com Probabilidade

A probabilidade é dada por:

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

👉 O PFC nos ajuda a encontrar o **total de casos possíveis**.

---



## 📌 Exemplo aplicado

Qual a probabilidade de tirar **cara e depois o número 6**?

- Total de possibilidades: 12  
- Casos favoráveis: 1  

$$
P = \frac{1}{12}
$$

---





```python
print(f"A probabilidade é de {(1/12)*100}%")
```

## 1. Conceito fundamental

Probabilidade mede a chance de um evento acontecer:

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

Exemplo:
Dado → pares = {2,4,6}

P(par) = 3/6 = 0.5


```python
# exemplo simples
pares = [2,4,6]
dado = [1,2,3,4,5,6]

probabilidade = len(pares)/len(dado)
print(f"A probabilidade de sair um número par é de {probabilidade *100}%")
```


## 5. Eventos dependentes (intuição)

Se retirar carta sem reposição → muda a probabilidade

Isso é MUITO importante para IA (dados mudam!)


EXEMPLO 1: CARTAS

“Se eu tiro uma carta e NÃO devolvo… o baralho muda.”

```python
cartas = ["❤️", "❤️", "♠", "♠"]
```

Qual a probabilidade de tirar duas cartas ❤️ seguidas sem reposição?

1ª carta:

* 2/4 = 1/2

2ª carta:

agora só sobram 3 cartas
apenas 1 é ❤️

* 1/3

P = 1/2 * 1/3 = 1/6

```python
import random

cartas = ["❤️", "❤️", "♠", "♠"]

sucesso = 0
n = 10000
# quantas vezes ele vai repetir o for

for _ in range(n):
    baralho = cartas.copy() # copy é pra ela não mudar
    c1 = random.choice(baralho) # minha primeira pegada
    baralho.remove(c1) # tiro essa carta do meu baralho

    c2 = random.choice(baralho) # baralho reduzido

    if c1 == "❤️" and c2 == "❤️": #vendo se eu tirei o coração nas 2 tiradas
        sucesso += 1

probabilidade = sucesso / n
print(sucesso)
print(f"A probabilidade de sair 2 cartas iguais seguidas é de {probabilidade * 100:.2f}% de vezes")
```

EXEMPLO 2 — ALUNOS

“Imagina que eu tenho uma turma com 3 alunos bons em Python e 2 iniciantes.”

Qual a probabilidade de escolher 2 alunos bons seguidos sem reposição?

1º aluno:

* 3/5

2º aluno:

agora só restam 4 alunos
2 são bons

* 2/4 = 1/2

```python
# P = (3/5) * (1/2) = 3/10
prob = (3/5) * (1/2)
print(f"A probabilidade de escolher 2 alunos bons seguidos sem reposição é de {prob *100}%")
```

```python
alunos = ["bom", "bom", "bom", "iniciante", "iniciante"]
```

```python
sucesso = 0
n = 10000

for _ in range(n):
    turma = alunos.copy()

    a1 = random.choice(turma)
    turma.remove(a1)

    a2 = random.choice(turma)

    if a1 == "bom" and a2 == "bom":
        sucesso += 1

sucesso / n
probabilidade = sucesso / n
print(sucesso)
print(f"A probabilidade de escolher 2 alunos bons seguidos sem reposição é de {probabilidade * 100 :.2f} de vezes")
```


## 6. Frequência vs probabilidade

Vamos simular:


```python

import random

n=10000
caras=0

for _ in range(n):
    if random.choice(["cara","coroa"])=="cara":
        caras+=1

probabilidade = caras/n
print(caras)
print(f"A probabilidade de sair cara é de {probabilidade * 100 :.2f} de vezes")
```

## 7. Probabilidade em IA

IA retorna algo assim:

```python
#Código
import numpy as np
prob = np.array([0.1, 0.7, 0.2])

classes = ["gato", "cachorro","onca"]

classes[np.argmax(prob)]
```


## 11. Desafio IA

Crie um sistema:

Entrada: probabilidades  
Saída: decisão da IA


```python
def decidir(probabilidades):
    return max(probabilidades, key=probabilidades.get)

dados = {
    "gato": 0.8,
    "cachorro": 0.2
}
resultado = decidir(dados)
print("Decisão da IA:", resultado)
```
