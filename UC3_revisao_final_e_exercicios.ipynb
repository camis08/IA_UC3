## 1. Função Linear

### Forma geral

$$
f(x) = ax + b
$$

### Ideia principal
A função linear representa uma relação de crescimento ou decrescimento constante.

- `a` = taxa de variação
- `b` = valor inicial

### Exemplo
$$
f(x) = 2x + 3
$$

Se `x = 4`:

$$
f(4) = 2 \cdot 4 + 3 = 11
$$

### Onde isso aparece em IA?
- regressão linear;
- previsão de valores;
- interpretação de relação entre entrada e saída.

```python
# Exemplo de função linear
def f_linear(x):
    return 2 * x + 3

print(f_linear(4))
```

```python
valores = [0,1,2,3,4]
for x in valores:
    print(f"f({x}) -> {f_linear(x)}")
```

## 2. Função Polinomial

### Exemplo mais importante na UC3
Função do 2º grau:

$$
f(x) = ax^2 + bx + c
$$

### Ideia principal
A função polinomial permite representar comportamentos curvos.

### Exemplo
$$
f(x) = x^2 - 4x + 5
$$

### Observações
- se `a > 0`, a parábola abre para cima;
- se `a < 0`, a parábola abre para baixo.

### Onde isso aparece em IA?
- modelagem de relações não lineares;
- transformação de variáveis (`x²`, `x³`);
- regressão polinomial.

```python
"""
Na modelagem de relações não lineares, na criação de variáveis polinomiais (como x²) e na regressão polinomial.
Ele determina o formato da curva: se a > 0, há um ponto mínimo (curva em U), e se a < 0, há um ponto máximo (curva invertida).
Isso é importante para entender comportamentos como otimização, erro de modelos e overfitting.
"""
```

```python
def desempenho(x):
    return x**2 - 4*x + 5

parametros = [0, 1, 2, 3, 4, 5, 6]
[(x, desempenho(x)) for x in parametros]
#print(f"f({x}) -> {desempenho(x)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 6, 100)
altura = x**2 - (4*x) + 5

plt.figure(figsize=(8, 5))
plt.plot(x,altura)
plt.title("Trajetória")
plt.xlabel("Parâmetro")
plt.ylabel("Desempenho")
plt.grid()
plt.show()
```

## 3. Função Exponencial

### Forma geral

$$
f(x) = a \cdot b^x
$$

### Ideia principal
Na função exponencial, o crescimento ou decrescimento acontece por multiplicação constante.

### Exemplo
$$
f(x) = 3 \cdot 2^x
$$

### Observações
- se `b > 1`, há crescimento;
- se `0 < b < 1`, há decrescimento.

### Onde isso aparece em IA?
- função sigmoid;
- probabilidades;
- crescimento de combinações;
- comportamento de certos algoritmos.

```python
def f_expo(x):
    return 3 * (2 ** x)

etapas = [0, 1, 2, 3, 4, 5]
[(x, f_expo(x)) for x in etapas]
```

- complexidade de algoritmos.

## 5. Probabilidade Básica

### Fórmula

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

### Ideias principais
- probabilidade mede chance;
- eventos independentes: multiplicamos probabilidades;
- “e” é diferente de “ou”.

### Exemplos
- probabilidade de cara em uma moeda: \(1/2\)
- probabilidade de cara duas vezes seguidas: \((1/2) \cdot (1/2) = 1/4\)

### Onde isso aparece em IA?
- classificação;
- decisão baseada em probabilidades;
- Naive Bayes;
- previsão com incerteza.

```python
# Exemplo simples de decisão da IA com probabilidades

classes = ["A", "B", "C"]
probabilidades = [0.2, 0.6, 0.2]

maior_prob = probabilidades[0]
indice_maior = 0

for i in range(len(probabilidades)):
    if probabilidades[i] > maior_prob:
        maior_prob = probabilidades[i]
        indice_maior = i

decisao = classes[indice_maior]

print("Probabilidades:", probabilidades)
print("Decisão da IA:", decisao)
```
